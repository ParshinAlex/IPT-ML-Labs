```python
import pandas as pd
import numpy as np
from sklearn import linear_model, preprocessing
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt

df = pd.read_csv('data/cars_dataset.csv')
models = dict(A1=1, A2=2, A3=3, A4=4, A5=5, A6=6, A7=7, A8=8, Q1=9, Q2=10, Q3=11, Q4=12, Q5=13, Q6=14, Q7=15, Q8=16,
              RS3=17, RS4=18, RS5=19, RS6=20, S3=21, S4=22, S5=23, S6=24, S7=25, S8=26, SQ5=27, SQ7=28, RS7=29, R8=30,
              TT=31)
```

```python
def prepare_data(data):
    data = data.replace(to_replace='Manual', value=0)
    data = data.replace(to_replace='Automatic', value=1)
    data = data.replace(to_replace='Petrol', value=0)
    data = data.replace(to_replace='Diesel', value=1)
    data = data.replace(to_replace='Electric', value=2)
    data = data[~data['transmission'].isin(['Semi-Auto', 'Other'])]
    data = data[~data['fuelType'].isin(['Hybrid', 'Other'])]
    data = data[data['Make'] == 'audi']
    data.drop('Make', axis=1, inplace=True)
    data = data.drop(data[data['mileage'] > 300000].index)
    data['model'] = data['model'].map(models)

    x = data.values
    min_max_scaler = preprocessing.MinMaxScaler()
    x_scaled = min_max_scaler.fit_transform(x)
    data_transformed = pd.DataFrame(x_scaled)
    data_transformed.columns = list(data.columns)
    
    
    
    print(data_transformed)

    return data_transformed


df = prepare_data(df)
```

```python
def simple_linear_regression(data, dots):
    inputs = data.head(len(data) // 2)["mileage"].to_numpy().reshape(-1, 1)
    outputs = data.head(len(data) // 2)["price"].to_numpy()

    
    model = linear_model.LinearRegression().fit(inputs, outputs)
    r_sq = model.score(inputs, outputs)

    print('coefficient of determination:', r_sq)
    print('intercept:', model.intercept_)
    print('slope:', model.coef_)

    predicted = model.predict(data.tail(len(data) // 2)["mileage"].to_numpy().reshape(-1, 1))
    res_predicted = pd.Series(predicted).reset_index(drop=True)
    res_real = data.tail(len(data) // 2)["price"].reset_index(drop=True)

    res = pd.DataFrame({'predicted_price': res_predicted, 'real_price': res_real})
    
    for i in range(len(res)):
        if (abs(res_predicted[i] - res_real[i]) > 0.1):
            res = res.drop(i)
    print(res.describe())
    
    x_axis = data.tail(len(data) // 2)["mileage"]

    plt.plot(x_axis.head(dots), res.head(dots)['predicted_price'], '.')
    plt.plot(x_axis.head(dots), res.head(dots)['real_price'], '.')
    plt.legend(res, loc=2)
    plt.xlim(0.0, 0.6)
    plt.ylim(0.0, 0.4)
    plt.show()
    
simple_linear_regression(df, 1000)
```

```python
real_price = df.tail(len(df) // 2)["price"].reset_index(drop=True)

def prepare_models(data):
    
    inputs = data.drop(labels='price', axis=1).head(len(data) // 2).to_numpy()
    outputs = data['price'].head(len(data) // 2).to_numpy()
    
    # Ordinary Linear Model
    
    ordinary = linear_model.LinearRegression()
    ordinary.fit(inputs, outputs)
    ordinary_sq = ordinary.score(inputs, outputs)
    
    print('Ordinary coefficient of determination: ', ordinary_sq)
    print('Ordinary intercept: ', ordinary.intercept_)
    print('Ordinary slope: ', ordinary.coef_, '\n')
    
    ordinary_predicted_price = ordinary.predict(data.drop(labels='price', axis=1).tail(len(data) // 2).to_numpy())
    ordinary_res_predicted = pd.Series(ordinary_predicted_price).reset_index(drop=True)
    ordinary_res = pd.DataFrame({'predicted_price': ordinary_res_predicted, 'real_price': real_price})
    
    # Ridge Model
    
    rid = linear_model.Ridge(alpha=0.5)
    rid.fit(inputs, outputs)
    rid_sq = rid.score(inputs, outputs)
    
    print('Ridge coefficient of determination: ', rid_sq)
    print('Ridge intercept: ', rid.intercept_)
    print('Ridge slope: ', rid.coef_, '\n')

    rid_predicted_price = rid.predict(data.drop(labels='price', axis=1).tail(len(data) // 2).to_numpy())
    rid_res_predicted = pd.Series(rid_predicted_price).reset_index(drop=True)
    rid_res = pd.DataFrame({'predicted_price': rid_res_predicted, 'real_price': real_price})
    
    # Lasso Model
    
    lasso = linear_model.Lasso(alpha=0.0005)
    lasso.fit(inputs, outputs)
    lasso_sq = lasso.score(inputs, outputs)
    
    print('Lasso coefficient of determination: ', lasso_sq)
    print('Lasso intercept: ', lasso.intercept_)
    print('Lasso slope: ', lasso.coef_, '\n')

    lasso_predicted_price = lasso.predict(data.drop(labels='price', axis=1).tail(len(data) // 2).to_numpy())
    lasso_res_predicted = pd.Series(lasso_predicted_price).reset_index(drop=True)
    lasso_res = pd.DataFrame({'predicted_price': lasso_res_predicted, 'real_price': real_price})
    
    # El-Net Model
    
    el_net = linear_model.ElasticNet(alpha=0.0005)
    el_net.fit(inputs, outputs)
    el_net_sq = el_net.score(inputs, outputs)
    
    print('Elastic-Net coefficient of determination: ', el_net_sq)
    print('Elastic-Net intercept: ', el_net.intercept_)
    print('Elastic-Net slope: ', el_net.coef_, '\n')

    el_net_predicted_price = el_net.predict(data.drop(labels='price', axis=1).tail(len(data) // 2).to_numpy())
    el_net_res_predicted = pd.Series(el_net_predicted_price).reset_index(drop=True)
    el_net_res = pd.DataFrame({'predicted_price': el_net_res_predicted, 'real_price': real_price})
    
    return [ordinary_res, rid_res, lasso_res, el_net_res]

models_predictions = prepare_models(df)
```

```python
def ensemble(models_predictions):
    fin_predicted_prices = [i for i in range(len(real_price))]
    temp1 = models_predictions[0]['predicted_price']
    temp2 = models_predictions[1]['predicted_price']
    temp3 = models_predictions[2]['predicted_price']
    temp4 = models_predictions[3]['predicted_price']
    for i in range(0, len(real_price), 1):
        fin_predicted_prices[i] = (temp1[i] + temp2[i] + temp3[i] + temp4[i]) / 4.0
        
    fin_res = pd.DataFrame({'predicted_price': fin_predicted_prices, 'real_price': real_price})
    
    print('Ensemble coefficient of determination: ', r2_score(real_price, fin_predicted_prices))
    return fin_res


ensemble_predictions = ensemble(models_predictions)
```

```python
def show_plot(model_res, model_name, dots, index):
    
    x_axis = np.arange(dots)

    plt.figure(figsize=(50, 10))
    plt.subplot(2, 3, index)
    plt.grid()
    plt.plot(x_axis, model_res.head(dots)['predicted_price'], 'r.')
    plt.plot(x_axis, model_res.head(dots)['real_price'], 'b.')
    plt.legend(model_res, loc=2)
    plt.title(model_name)
    

    
show_plot(models_predictions[0], 'Ordinary', 100, 1)
show_plot(models_predictions[1], 'Ridge', 100, 2)
show_plot(models_predictions[2], 'Lasso', 100, 3)
show_plot(models_predictions[3], 'El-Net', 100, 4)
show_plot(ensemble_predictions, 'Ensemble', 100, 5)
plt.show()
```
